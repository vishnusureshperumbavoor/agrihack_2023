```python
season_names=[]
```

```python
# this should run only one time
for tt in df["month_column"]:
    print(tt)
    if tt=="Jan" or tt=="Feb":
        season_names.append("Winter")
    elif tt=="Mar" or tt=="Apr":
        season_names.append("Spring")
    elif tt=="May" or tt=="June":
        season_names.append("Summer")
    elif tt=="July" or tt=="Aug":
        season_names.append("Monsoon")
    elif tt=="Sep" or tt=="Oct":
        season_names.append("Autumn")
    elif tt=="Nov" or tt=="Dec":
        season_names.append("Pre Winter")
```

```python
len(season_names)
```

```python
df["season_names"] = season_names
```
